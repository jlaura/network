```python
def compute_length(v0, v1):
    return math.sqrt((v0[0] - v1[0])**2 + (v0[1] - v1[1])**2)

shps = ps.open('geodanet/streets.shp', 'r')
adjacencylist = defaultdict(list)
nodes = {}
edge_lengths = {}
nodecount = 0
for shp in shps:
    vertices = shp.vertices
    for i, v in enumerate(vertices[:-1]):
        try:
            vid = nodes[v]
        except:
            nodes[v] = vid = nodecount
            nodecount += 1
        try:
            nvid = nodes[vertices[i+1]]
        except:
            nodes[vertices[i+1]] = nvid = nodecount
            nodecount += 1
            
        adjacencylist[vid].append(nvid)
        adjacencylist[nvid].append(vid)
        
        edge = (vid, nvid)
        
        length = compute_length(v, vertices[i+1])
        edge_lengths[edge] = length
```
